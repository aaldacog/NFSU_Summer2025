```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import SGD
import pandas as pd
```

```python
# Load your dataset
df = pd.read_csv("data_Temperature_Humidity.csv")
X = df[['Temperature', 'Humidity']].values
y = LabelEncoder().fit_transform(df['Class']).reshape(-1, 1)
```

```python
# Normalize inputs
X = MinMaxScaler().fit_transform(X)
```

```python
# Define the model
model = Sequential([
    Dense(2, input_dim=2, activation='sigmoid'),  # Layer L1
    Dense(1, activation='sigmoid')                # Output Layer L2
])
```

```python
# Compile the model
model.compile(optimizer=SGD(learning_rate=0.1), loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# Train the model
model.fit(X, y, epochs=1000, verbose=1)
```

```python
# Predict
y_pred = model.predict(X[:8])
print(y_pred)
```
